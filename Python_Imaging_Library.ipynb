```python
from PIL import Image, ImageFilter, ImageFont, ImageDraw
from matplotlib.pyplot import imshow
```

```python
file_name = "pavilion.jpg"
im = Image.open(file_name)
imshow(im)
```

```python
print(im.size)
print(im.format_description)
print(im.mode)
```

```python
crop = (1000, 1000, 2000, 2000)
cropped_image = im.crop(crop)
imshow(cropped_image)
```

```python
resized_im = im.resize((1000, 800))
imshow(resized_im)
```

```python
gray_scale = im.convert('LA')
imshow(gray_scale)
```

```python
effect = im.filter(ImageFilter.GaussianBlur(radius=20))
imshow(effect)
```

```python
draw = ImageDraw.Draw(im)
font = ImageFont.truetype("arial.ttf", 400)
draw.text((3200, 3000), "Aditi' Photo!", font=font)
imshow(im)
```
